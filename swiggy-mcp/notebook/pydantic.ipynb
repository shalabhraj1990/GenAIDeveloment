```python
from pydantic import BaseModel,Field

class CalcInput(BaseModel):
    a: int
    b: int
    
model_input = CalcInput(a=3,b=8)
model_input
```

```python
#converting a model into dictionary
model_input.model_dump()
type(model_input.model_dump())
```

```python
#converting a model from dictionay
model_from_input = CalcInput(**input)
model_from_input
```

```python
## Pydantic Field Validators
#1 Basic validation
from pydantic import Field,EmailStr,HttpUrl
class FormData(BaseModel):
    username:str
    email:str
    password:str
    profile_page:str
    
#2 Field validtion [V1 - older version]
class FormData_V1(BaseModel):
    username:str = Field(min_length=6, max_length=20)
    email:EmailStr
    password:str
    profile_page:HttpUrl
```

```python
#Valid Data
raw_form = {
    "username":"administrator",
    "email":"admin@vt.com",
    "password":"passowrd@23423",
    "profile_page":"http://google.com"
}

validate_date = FormData_V1(**raw_form)
validate_date
```

```python
#INValid Data
raw_form = {
    "username":"administrator",
    "email":"admin@.com",
    "password":"passowrd@23423",
    "profile_page":"http://google.com"
}

validate_date = FormData_V1(**raw_form)
validate_date
```

```python
from pydantic import BaseModel,Field
from typing import Annotated,Literal

class Item(BaseModel):
    id:Annotated[int,Field(gt=0)]
    name:Annotated[str,Field(description="name of item")]
    qty: Annotated[int,Field(ge=1)]

class Order(BaseModel) :
    orderId : Annotated[str,Field(description="order id")]
    customerName:Annotated[str,Field(gt=0,description="customer name")]
    status :Annotated[Literal["delivered","processing","cancelled","not delivered"],Field(description="status of order")]
    items : Annotated[list[Item],Field(description="list of items")]
```

```python
row_data = {
	"orderId": "ORD_64829",
	"customerName": "Isha Kapoor",
	"status": "delivered",
	"items": [
		{
			"id": 3453,
			"qty": 1,
			"price": 89.0,
			"name": "books"
		}
	]
};
validation = Order(**row_data)
```
